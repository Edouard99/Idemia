```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dic_results=torch.load("./Results/checkpoint_30.pth",map_location=torch.device('cpu'))
```

```python
loss_train_train=dic_results["loss_train_train"]
loss_val_eval=dic_results["loss_val_eval"]
loss_train_eval=dic_results["loss_train_eval"]
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,5))
fig.suptitle('Loss value (BCE) of training and validation sets as a function of epoch', fontsize=16)
axs[0].plot(range(len(loss_train_eval)),loss_train_eval)
axs[0].set_title("Training set")
axs[0].set_xlabel("Epochs")
axs[0].set_ylabel("Loss value BCE")
axs[1].plot(range(len(loss_val_eval)),loss_val_eval)
axs[1].set_title("Validation set")
axs[1].set_xlabel("Epochs")
axs[1].set_ylabel("Loss value BCE")
```

```python
hter_training=[]
hter_validation=[]
for k in range(31):
    dic=torch.load("./Results/checkpoint_{}.pth".format(k),map_location=torch.device('cpu'))
    hter_training.append(dic['metrics_train']["hter"])
    hter_validation.append(dic['metrics_val']["hter"])
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,5))
fig.suptitle('HTER of training and validation sets as a function of epoch', fontsize=16)
axs[0].plot(range(len(hter_training)),hter_training)
axs[0].set_title("Training set")
axs[0].set_xlabel("Epochs")
axs[0].set_ylabel("HTER")
axs[1].plot(range(len(hter_validation)),hter_validation)
axs[1].set_title("Validation set | Min of Hter = {} | Argmin of HTER = Epoch{}".format(np.round(np.min(hter_validation),3),np.argmin(hter_validation)))
axs[1].set_xlabel("Epochs")
axs[1].set_ylabel("HTER")
```
